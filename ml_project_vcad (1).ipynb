#MEAN SQUARED ERROR

```python
# Function to compute Mean Squared Error (MSE) between two frames
def compute_mse(frame1, frame2):
    return np.mean(np.square(frame1 - frame2))

# Load images and calculate MSE between consecutive images
mse_values = []
for i in range(len(test_frames) - 1):
    # Load images
    frame1 = cv2.imread(os.path.join(frames_path,test_frames[i]))
    frame2 = cv2.imread(os.path.join(frames_path,test_frames[i + 1]))
    
    # Convert images to grayscale if needed
    frame1_gray = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    frame2_gray = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

    # Compute MSE between frames
    mse = compute_mse(frame1_gray, frame2_gray)
    mse_values.append(mse)
```

```python
# Example similarity values (replace this with your actual list of similarity values)

# Calculate the standard deviation of similarity values
std_dev = np.std(mse_values)

# Set a threshold (e.g., 3 times the standard deviation)
threshold = 3 * std_dev

# Find indices where significant changes occur based on the threshold
mse_values_indices = [i for i, value in enumerate(mse_values) if value > threshold]

print("Indices with significant changes:", mse_values_indices)
```

#Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

# Function to flatten and convert image to vector
def image_to_vector(image_path):
    img = cv2.imread(image_path)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    flattened_img = img_gray.flatten()
    return flattened_img.reshape(1, -1)  # Reshape to a row vector

# Convert images to vectors and compute cosine similarity
cosine_sim_values = []
for i in range(len(test_frames) - 1):
    # Convert images to vectors
    vector1 = image_to_vector(os.path.join(frames_path,test_frames[i]))
    vector2 = image_to_vector(os.path.join(frames_path,test_frames[i+1]))
    
    # Compute cosine similarity between vectors
    cosine_sim = cosine_similarity(vector1, vector2)
    cosine_sim_values.append(cosine_sim[0][0])
```

```python
cosine_sim_values[100:200]
```

```python
# Example similarity values (replace this with your actual list of similarity values)

# Calculate the standard deviation of similarity values
std_dev = np.std(cosine_sim_values)

# Set a threshold (e.g., 3 times the standard deviation)
threshold = 3 * std_dev

# Find indices where significant changes occur based on the threshold
cosine_sim_values_indices = [i for i, value in enumerate(cosine_sim_values) if value > threshold]

print("Indices with significant changes:", cosine_sim_values_indices)
```
