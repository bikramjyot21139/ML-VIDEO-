```python
from moviepy.editor import VideoFileClip

def extract_audio(input_video, output_audio):
    video_clip = VideoFileClip(input_video)
    audio_clip = video_clip.audio
    audio_clip.write_audiofile(output_audio)

# Example usage
input_video_file = "Ave.mkv"
output_audio_file = "output_audio.mp3"
extract_audio(input_video_file, output_audio_file)
```

```python
from pydub import AudioSegment

def extract_first_minute(input_audio, output_first_minute):
    audio = AudioSegment.from_file(input_audio)
    first_minute = audio[:60000]  # Extract the first minute (in milliseconds)
    first_minute.export(output_first_minute, format="wav")

# Example usage
input_audio_file = "output_audio.wav"
output_first_minute_file = "output1.mp3"
extract_first_minute(input_audio_file, output_first_minute_file)
```

```python
import numpy as np
from sklearn.decomposition import PCA
import librosa
import soundfile as sf

# Load audio data
file_path = "output_audio.mp3"
audio_data, sr = librosa.load(file_path, sr=None)

# Reshape the audio data into a 2D array (features x time steps)
audio_data_2d = audio_data.reshape(1, -1)

# Choose the number of principal components (compression size)
compression_size = 0.3  # Adjust based on the number of features
pca = PCA(n_components=compression_size)

# Fit the PCA model on the audio data
compressed_representation1 = pca.fit_transform(audio_data_2d)

# Save the compressed representation and metadata
np.save("compressed_representation1.npy", compressed_representation1)

```

```python
# Decompress audio
reconstructed_audio_2d = pca.inverse_transform(compressed_representation1)
reconstructed_audio = reconstructed_audio_2d.flatten()

# Save the reconstructed audio
sf.write("reconstructed_audio2.mp3", reconstructed_audio, sr)
```

```python
from moviepy.video.io.VideoFileClip import VideoFileClip
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_audio

def remove_audio(input_video_path, output_video_path):
    # Load the video clip
    video_clip = VideoFileClip(input_video_path)

    # Remove audio from the video
    video_clip = video_clip.without_audio()

    # Save the video without audio
    video_clip.write_videofile(output_video_path, codec="libx264", audio_codec="aac", temp_audiofile="temp_audio.aac", remove_temp=True)


input_video_path = "Ave.mkv"
output_video_path = "output_video_no_audio.mp4"

remove_audio(input_video_path, output_video_path)
```
